# As of my last update in September 2021, there is no official term or specific category known as "High_Taxpayers" recognized by tax authorities. However, it is common to refer to individuals or businesses with higher income or significant financial resources as high-income taxpayers.

High-income taxpayers typically decide on their tax planning and strategies based on a combination of factors, including their income sources, deductions, investments, and the tax laws and regulations of their country or state. Here are some common ways high-income taxpayers might make decisions regarding their taxes:

Tax Professionals: High-income taxpayers often seek the guidance of tax professionals, such as certified public accountants (CPAs) or tax advisors, who specialize in helping clients navigate complex tax laws and optimize their tax situations.

Income Structuring: They may consider structuring their income in a way that takes advantage of favorable tax treatments for certain types of earnings or investments. This could involve diversifying income streams, managing capital gains, and strategically timing the realization of income.

Deductions and Credits: High-income taxpayers might explore eligible deductions and tax credits to reduce their taxable income. These could include charitable contributions, mortgage interest deductions, education credits, and more.

Investment Choices: Tax efficiency is a critical consideration for high-income taxpayers. They may choose investments with tax advantages, such as tax-exempt municipal bonds, retirement accounts, or tax-advantaged savings accounts.

Estate Planning: High-income taxpayers may engage in estate planning to minimize estate taxes and ensure a smooth transfer of wealth to their heirs.

Business Structures: For high-income individuals who own businesses, selecting the appropriate business structure (e.g., sole proprietorship, partnership, corporation) can have significant tax implications.

Tax Avoidance vs. Tax Evasion: High-income taxpayers must be careful to engage in legal tax planning and avoidance strategies, as opposed to illegal tax evasion, which involves intentionally underreporting income or falsifying records to evade taxes.

It is important to note that tax laws can change over time, and strategies that were beneficial in the past may not be as advantageous in the future. High-income taxpayers often need to stay informed about changes in tax laws and regulations to adjust their tax planning accordingly.

# Hard voting classification

In hard voting (also known as majority voting), every individual classifier votes for a class, and the majority wins. In statistical terms, the predicted target label of the ensemble is the mode of the distribution of individually predicted labels.

```python
# from sklearn.neighbors import KNeighborsClassifier
# from sklearn.ensemble import VotingClassifier
# from sklearn.linear_model import LogisticRegression
# from sklearn.svm import SVC
# from sklearn.tree import DecisionTreeClassifier

# lr_hard = LogisticRegression()
# lr_hard.fit(Train_X, Train_Y)
# knn_hard = KNeighborsClassifier(7)
# knn_hard.fit(Train_X, Train_Y)
# svc_hard = SVC(C = 10, probability = True)
# svc_hard.fit(Train_X, Train_Y)

# voting_clf_hard = VotingClassifier(estimators=[('lr', lr_hard), ('knn', knn_hard), ('svc', svc_hard)], voting='hard')
# voting_clf_hard.fit(Train_X, Train_Y)

# from sklearn.metrics import accuracy_score
# for clf in (lr_hard, knn_hard, svc_hard, voting_clf_hard):
#     clf.fit(Train_X, Train_Y)
#     y_test_pred = clf.predict(Test_X)
#     y_train_pred = clf.predict(Train_X)
#     print(clf.__class__.__name__, accuracy_score(Test_Y, y_test_pred))




from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# Instantiate the classifiers
lr_hard = LogisticRegression()
lr_hard.fit(Train_X, Train_Y)

knn_hard = KNeighborsClassifier(7)
knn_hard.fit(Train_X, Train_Y)

svc_hard = SVC(C=10, probability=True)
svc_hard.fit(Train_X, Train_Y)

dt_hard = DecisionTreeClassifier()
dt_hard.fit(Train_X, Train_Y)

knn2_hard = KNeighborsClassifier(5)
knn2_hard.fit(Train_X, Train_Y)

# Create the VotingClassifier with 'hard' voting
voting_clf_hard = VotingClassifier(
    estimators=[
        ('lr', lr_hard),
        ('knn', knn_hard),
        ('svc', svc_hard),
        ('dt', dt_hard),
        ('knn2', knn2_hard)
    ],
    voting='hard'
)

# Fit the VotingClassifier on the training data
voting_clf_hard.fit(Train_X, Train_Y)

from sklearn.metrics import accuracy_score

# Evaluate the individual classifiers and the ensemble
for clf in (lr_hard, knn_hard, svc_hard, dt_hard, knn2_hard, voting_clf_hard):
    y_test_pred = clf.predict(Test_X)
    y_train_pred = clf.predict(Train_X)
    print(clf.__class__.__name__, "Train accuracy:", accuracy_score(Train_Y, y_train_pred))
    print(clf.__class__.__name__, "Test accuracy:", accuracy_score(Test_Y, y_test_pred))
```
